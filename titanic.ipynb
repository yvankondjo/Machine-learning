```python
#on importe nos donnees
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
#on lit nos donnes
train=pd.read_csv('train.csv')
test=pd.read_csv('eval.csv')

```

```python
#on prepare nos donnees
y_train=train['survived']
x_train=train.drop(['survived'],axis=1)
y_test=test['survived']
x_test=test.drop(['survived'],axis=1)
```

```python
def function_pipeline():
    #on classe nos donnes par categories
    numerical_features=['age','fare']
    categorical_features=['sex','class','deck','embark_town','alone']
    #on creer une chaine de transformation pour nos donnees numerique en supprimant les lignes manquantes et 
    #en appliquant la normalisation standardscaler de tel sorte que nos donnes on une moyenne nul et un ecart type de 1
    numerical_pipeline=make_pipeline(SimpleImputer(),StandardScaler())
    #on creer une chaine de transformation  pour nos donnees categorielle en remplacant les donnees manquantes pas 
    #les plus frequent puis on les encode de la maniere oneHot
    categorical_pipeline=make_pipeline(SimpleImputer(strategy='most_frequent'),OneHotEncoder())
    #on creer maintenant un transformer qui s'applique uniquement a certaines donnes
    trans=make_column_transformer((numerical_pipeline,numerical_features),(categorical_pipeline,categorical_features))
    #maintenant on creer notre pipeline finale qui prends en parametre notre tansformer et notre estimateur 
    pipf=make_pipeline(trans,KNeighborsClassifier())
    return(pipf)
```

```python
#on recupere notre pipeline
model=function_pipeline()
#on parametre les parametre de notre model
params={'kneighborsclassifier__n_neighbors':range(1,20),'kneighborsclassifier__metric':['manhattan','euclidian']}
#on utilise la methode grid search cv
grid=GridSearchCV(model,param_grid=params,cv=5)
```

```python
#on teste notre model
grid.fit(x_train,y_train)

```

```python
grid.best_estimator_.score(x_test,y_test)
```
